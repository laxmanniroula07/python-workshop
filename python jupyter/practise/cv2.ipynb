```python
import cv2
import matplotlib.pyplot as plt
```

```python
config_files='ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model='frozen_inference_graph.pb'
```

```python
model=cv2.dnn_DetectionModel(frozen_model,config_files)
```

```python
classLabels=[]
file_name='labels.txt'
with open(file_name,'rt')as fpt:
   classLabels=fpt.read().rstrip('\n').split('\n')
```

```python
print(len(classLabels))
print(classLabels)
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)
model.setInputScale((127.5,127.5,127.5))
model.setInputSwapRB(True)
```

```python
img=cv2.imread('pp.jpeg')
plt.imshow(img)
```

```python
ClassIndex,confidence,bbox=model.detect(img, confThreshold=0.5)
```

```python
print(ClassIndex)
```

```python
font_scale=3
font=cv2.FONT_HERSHEY_PLAIN
for ClassInd,conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
    cv2.rectangle(img,boxes,(255,0,0),3)
    cv2.putText(img, classLabels[ClassInd-1],(boxes[0]+10,boxes[1]+40),  font, fontScale=font_scale,color=(55,2,0))
```
