```python
#Exercise 1: We will start with a simple problem that can be represented using the following graph.

from collections import deque
def bfs(graph, start, goal):
    visited = set()
    queue = deque([(start, [])])
    while queue:
        node, path = queue.popleft()
        if node == goal:
            return path + [node]
        if node not in visited:
            visited.add(node)
            for neighbor in graph[node]:
                queue.append((neighbor, path + [node]))
    return None
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}
start_node = 'A'
goal_node = 'E'
result = bfs(graph, start_node, goal_node)
if result:
    print("Path found:", result)    
else:
    print("No path found.")
```

```python
#Q 2 : Use the time library to calculate how long would it take to solve the problem using three different starting points and ending points


from collections import deque
import time
def bfs(graph, start, goal):
    visited = set()
    queue = deque([(start, [])])
    while queue:
        node, path = queue.popleft()
        if node == goal:
            return path + [node]
        if node not in visited:
            visited.add(node)
            for neighbor in graph[node]:
                queue.append((neighbor, path + [node]))
    return None
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Define starting points and ending points for three different scenarios
scenarios = [
    ('A', 'F'),
    ('B', 'C'),
    ('C', 'E')
]

# Measure time for each scenario
for start_node, goal_node in scenarios:
    start_time = time.time()
    result = bfs(graph, start_node, goal_node)
    end_time = time.time()

    if result:
        print(f"Path found from {start_node} to {goal_node}: {result}")
    else:
        print(f"No path found from {start_node} to {goal_node}.")

    time_taken = end_time - start_time
    print(f"Time taken for scenario ({start_node}, {goal_node}): {time_taken} seconds\n")
```

```python
#Q 3: Change the script to mode the following graph:
import heapq
import matplotlib.pyplot as plt
import networkx as nx

class Node:
    def __init__(self, name, heuristic):
        self.name = name
        self.heuristic = heuristic
        self.g_cost = float('inf')
        self.f_cost = float('inf')
        self.adjacent = {}
        self.parent = None

    def add_neighbor(self, neighbor, cost):
        self.adjacent[neighbor] = cost

    def __lt__(self, other):
        return self.f_cost < other.f_cost

def astar_search(start, goal):
    open_list = []
    closed_set = set()
    start.g_cost = 0
    start.f_cost = start.heuristic
    heapq.heappush(open_list, start)

    while open_list:
        current_node = heapq.heappop(open_list)

        if current_node == goal:
            path = []
            while current_node is not None:
                path.append(current_node.name)
                current_node = current_node.parent
            return path[::-1]

        closed_set.add(current_node)

        for neighbor, cost in current_node.adjacent.items():
            tentative_g_cost = current_node.g_cost + cost

            if neighbor in closed_set and tentative_g_cost >= neighbor.g_cost:
                continue

            if tentative_g_cost < neighbor.g_cost or neighbor not in open_list:
                neighbor.g_cost = tentative_g_cost
                neighbor.f_cost = tentative_g_cost + neighbor.heuristic
                neighbor.parent = current_node
                if neighbor not in open_list:
                    heapq.heappush(open_list, neighbor)

    return None

# Creating nodes
A = Node("A", 5)
B = Node("B", 4)
C = Node("C", 3)
D = Node("D", 2)
E = Node("E", 1)
F = Node("F", 0)
G = Node("G", 2)
H = Node("H", 3)

# Adding neighbors and their costs
A.add_neighbor(B, 1)
A.add_neighbor(C, 2)
B.add_neighbor(D, 3)
B.add_neighbor(E, 1)
C.add_neighbor(F, 3)
D.add_neighbor(G, 2)
E.add_neighbor(F, 2)
F.add_neighbor(H, 1)
G.add_neighbor(H, 3)

# Visualization of the graph
G = nx.Graph()
edges = [
    (A.name, B.name, {'weight': 1}),
    (A.name, C.name, {'weight': 2}),
    (B.name, D.name, {'weight': 3}),
    (B.name, E.name, {'weight': 1}),
    (C.name, F.name, {'weight': 3}),
    (D.name, G.name, {'weight': 2}),
    (E.name, F.name, {'weight': 2}),
    (F.name, H.name, {'weight': 1}),
    (G.name, H.name, {'weight': 3})
]
G.add_edges_from(edges)
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos, with_labels=True, node_size=3000, node_color='skyblue')
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title('Nodes Graph')
plt.show()

# Running the A* search algorithm
path = astar_search(A, H)
print("A* path:", path)
```

```python
from collections import deque

def bfs_maze(maze, start, goal):
    rows, cols = len(maze), len(maze[0])
    visited = set()
    parent = {} # To keep track of the parent of each cell in the path
    
    def is_valid(x, y):
        return 0 <= x < rows and 0 <= y < cols and maze[x][y] != '#' and (x, y) not in visited

    queue = deque([start])
    visited.add(start)
    found = False
    
    while queue:
        x, y = queue.popleft()
        if (x, y) == goal: 
            found = True
            break
        
        # Define possible moves: up, down, left, right
        moves = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        for dx, dy in moves:
            new_x, new_y = x + dx, y + dy
            if is_valid(new_x, new_y):
                queue.append((new_x, new_y))
                visited.add((new_x, new_y))
                parent[(new_x, new_y)] = (x, y)
    
    if not found: 
        return "No path found"
    
    # Reconstruct the path from the goal to the start
    path = []
    x, y = goal
    while (x, y) != start:
        path.append((x, y))
        x, y = parent[(x, y)]
    path.append(start)
    path.reverse()
    
    # Update the maze with the path
    for x, y in path:
        if maze[x][y] != 'S' and maze[x][y] != 'G':
            maze[x][y] = '*'
    
    return "Path found"

# Define the maze
maze = [
    ['S', '.', '.', '.', '.', '.', '#', '.', '.'],
    ['#', '#', '#', '#', '#', '.', '#', '#', '#'],
    ['.', '.', '.', '.', '.', '.', '.', '.', 'G'],
    ['#', '#', '#', '#', '#', '#', '#', '#', '#']
]

# Starting and goal positions
start_position = (0, 0)
goal_position = (2, 8)

result = bfs_maze(maze, start_position, goal_position)

# Print the maze with the path
for row in maze:
    print(' '.join(row))
print(result)
```

```python
#Q: Try different starting and ending points then calculate the number of steps needed to solve the problem.
from collections import deque

def bfs_maze(maze, start, goal):
    rows, cols = len(maze), len(maze[0])
    visited = set()
    parent = {} # To keep track of the parent of each cell in the path
    
    def is_valid(x, y):
        return 0 <= x < rows and 0 <= y < cols and maze[x][y] != '#' and (x, y) not in visited

    queue = deque([(start, 0)]) # Queue stores tuples of (position, step_count)
    visited.add(start)
    found = False
    steps = 0
    
    while queue:
        (x, y), steps = queue.popleft()
        if (x, y) == goal: 
            found = True
            break
        
        # Define possible moves: up, down, left, right
        moves = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        for dx, dy in moves:
            new_x, new_y = x + dx, y + dy
            if is_valid(new_x, new_y):
                queue.append(((new_x, new_y), steps + 1))
                visited.add((new_x, new_y))
                parent[(new_x, new_y)] = (x, y)
    
    if not found: 
        return "No path found", -1
    
    # Reconstruct the path from the goal to the start
    path = []
    x, y = goal
    while (x, y) != start:
        path.append((x, y))
        x, y = parent[(x, y)]
    path.append(start)
    path.reverse()
    
    # Update the maze with the path
    for x, y in path:
        if maze[x][y] != 'S' and maze[x][y] != 'G':
            maze[x][y] = '*'
    
    return "Path found", steps

# Define the maze
maze = [
    ['S', '.', '.', '.', '.', '.', '#', '.', '.'],
    ['#', '#', '#', '#', '#', '.', '#', '#', '#'],
    ['.', '.', '.', '.', '.', '.', '.', '.', 'G'],
    ['#', '#', '#', '#', '#', '#', '#', '#', '#']
]

# Function to test different start and goal positions
def test_maze(maze, start, goal):
    # Make a deep copy of the maze to avoid modifying the original
    from copy import deepcopy
    test_maze = deepcopy(maze)
    
    # Update start and goal positions in the maze
    test_maze[start[0]][start[1]] = 'S'
    test_maze[goal[0]][goal[1]] = 'G'
    
    result, steps = bfs_maze(test_maze, start, goal)
    
    # Print the maze with the path
    for row in test_maze:
        print(' '.join(row))
    print(result)
    print(f"Number of steps: {steps}")

# Test different start and goal positions
start_position = (0, 0)
goal_position = (2, 8)
print("Test 1:")
test_maze(maze, start_position, goal_position)
print()

start_position = (0, 1)
goal_position = (2, 7)
print("Test 2:")
test_maze(maze, start_position, goal_position)
print()

start_position = (2, 0)
goal_position = (2, 8)
print("Test 3:")
test_maze(maze, start_position, goal_position)
print()

start_position = (2, 1)
goal_position = (0, 8)
print("Test 4:")
test_maze(maze, start_position, goal_position)
```
