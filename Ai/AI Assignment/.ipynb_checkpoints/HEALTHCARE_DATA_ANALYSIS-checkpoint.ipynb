```python
 def summary(healthcare):
    print(f'data shape: {healthcare.shape}')
    summ = pd.DataFrame(healthcare.dtypes, columns=['data type'])
    summ['#missing'] = healthcare.isnull().sum().values 
    summ['%missing'] = healthcare.isnull().sum().values / len(healthcare) * 100
    summ['#unique'] = healthcare.nunique().values
    desc = pd.DataFrame(healthcare.describe(include='all').transpose())
    summ['top'] = desc['top'].values
    summ['freq'] = desc['freq'].values
    summ['min'] = desc['min'].values
    summ['max'] = desc['max'].values
    summ['average'] = desc['mean'].values
    summ['standard_deviation'] = desc['std'].values
    summ['first value'] = healthcare.loc[0].values
    summ['second value'] = healthcare.loc[1].values
    summ['third value'] = healthcare.loc[2].values
    
    return summ.style.background_gradient(cmap='YlOrBr')
```

```python
summary(healthcare)
```

```python


# Convert the 'Date of Admission' and 'Discharge Date' columns to datetime, coercing errors to NaT
healthcare['Date of Admission'] = pd.to_datetime(healthcare['Date of Admission'], errors='coerce')
healthcare['Discharge Date'] = pd.to_datetime(healthcare['Discharge Date'], errors='coerce')

# Initialize the 'Duration' column with NaN
healthcare['Duration'] = np.nan
healthcare['Blood Group'] = np.nan

# Loop through the DataFrame and update 'Duration'
for i in range(len(healthcare)):
    date_of_admission = healthcare.loc[i, 'Date of Admission']
    discharge_date = healthcare.loc[i, 'Discharge Date']
    blood_type=healthcare.loc[i, 'Blood Group']
    
    if pd.notna(date_of_admission) and pd.notna(discharge_date):
        healthcare.loc[i, 'Duration'] = (discharge_date - date_of_admission).days
        
        
        if pd.notna(blood_type):  
            healthcare.loc[i, 'Blood Group'] = blood_type[:-1]

# Convert 'Duration' to integer
healthcare['Duration'] = healthcare['Duration'].astype('Int64')  # Using 'Int64' to allow for NaNs
```

```python
categorical_columns = ['Gender', 'Blood Type', 'Medical Condition']
numerical_columns = ['Age']
# Extract features (X) and target variable (y)
X = healthcare[categorical_columns + numerical_columns]
y = healthcare['Test Results']

# Create a column transformer for one-hot encoding
preprocessor = ColumnTransformer(            #cat= categor
    transformers=[
        ('cat', OneHotEncoder(drop='first'), categorical_columns,)
    ],
    remainder='passthrough'
)
# Apply one-hot encoding to X
X = preprocessor.fit_transform(X)
```

```python
x_for_understanding = pd.DataFrame(X.toarray(), columns=preprocessor.get_feature_names_out())
display(x_for_understanding.head())
x_for_understanding.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 3.00, random_state = 7)
```

```python
#  Decision Tree Classifier with criterion entropy
```

```python
classifier = DecisionTreeClassifier(criterion='entropy', splitter='best',max_depth=5,min_samples_leaf=8, random_state=40)
classifier.fit(X_train, y_train)
```

```python
# Predicting the test results
y_pred=classifier.predict(X_test)
y_pred
```

```python
#Check accuracy score with criterion entropy 
```

```python
y_pred_train =classifier.predict(X_train)
y_pred_train
```
